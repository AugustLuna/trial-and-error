```python

np.arange(0,50,1)

np.linspace(0,50,2)

x=np.arange(0,100) #plots it to 100
y=np.arange(0,100) #same thing

X,Y= np.meshgrid(x,y) #this makes the grid

plt.pcolormesh(Y) 
plt.colorbar()
plt.xlabel("Column index", fontsize='18')
plt.ylabel("Row index", fontsize='18')
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.arange(0,50,1)

np.linspace(0,50,2)

x=np.arange(0,100)
y=np.arange(0,100)

X,Y= np.meshgrid(x,y)

Z= np.cos(X*Y)

plt.pcolormesh(Z)
plt.colorbar()
plt.xlabel("Column index", fontsize='18')
plt.ylabel("Row index", fontsize='18')
```
